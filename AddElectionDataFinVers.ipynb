```python
import pandas as pd
import numpy as np
```

```python

election_list_df = []
```

```python
#To ensure that always the correct dataframe is used

def census_orig():
    census_orig_df = pd.read_pickle("ElecDistr2016Simp.pkl")
    census_orig_df['Results'] = 'unknown'
    return census_orig_df
    
```

```python
#Transforms the *.csv files with the electionresults so that they can be added to the dataframe

def partylist(electionresults):
    counter = electionresults.shape[0]
    pos = electionresults.shape[1] - 1
    results = {}
    i = 0
    NDP = 0
    Liberal = 0
    Bloc = 0
    Conservative = 0
    while i < counter:
        sample = electionresults.iloc[i]
        if "NDP" in sample[pos]:
            results[sample[2]] = "NDP"
            NDP += 1
        elif "Liberal" in sample[pos]:
            results[sample[2]] = "Liberal"
            Liberal += 1
        elif "Conservative" in sample[pos]: 
            results[sample[2]] = "Conservative"
            Conservative += 1
        elif "Bloc" in sample[pos]:
            results[sample[2]] = "Bloc"
            Bloc += 1
        elif "Green" in sample[pos]:
            results[sample[2]] = "Green"
        elif "Independent" in sample[pos]:
            results[sample[2]] = "Independent"
        elif "No Affiliation" in sample[pos]:
            results[sample[2]] = 'Independent'
        elif "N.D.P" in sample[pos]:
            results[sample[2]] = "NDP"
            NDP += 1
        else: print(sample, "not tracked")
        i += 1
        
    print("Conservative=", Conservative, "Liberal=", Liberal, "NDP= ", NDP, "Bloc=", Bloc)
    return results
```

```python
#Function for adding the election results to the census data

def resultsframe(listyear, listyear_df):
    counter = listyear_df.shape[0]
    no_data = []
    i = 0
    #print(listyear, listyear_df.head(), listyear_df.shape)
    while i < counter:
        distr = listyear_df.iloc[i][0]
        if int(distr) in listyear:
            party = listyear[int(distr)]
            listyear_df.at[i, 'Results'] = party
        else: 
            no_data.append(i)
        i += 1

    print(no_data)
    for district in no_data:
        listyear_df.drop([district], inplace=True)  

#Need to reset the index after the rows were dropped
    listyear_df.reset_index(inplace = True)

    listyear_df.drop(columns = ['index'], inplace = True)
    print(listyear_df.shape)

    return listyear_df
```

```python
#Function receives a list of results and counts the number of seats for each party.
#Results are returned as a dictionary.
#Does not work for Green and Independent.

def countResults(results):
  liberal = 0
  conservative = 0
  ndp = 0
  bloc = 0
  for i in range(0, len(results)-1):
    if results[i] == 'Conservative':
      conservative +=1
    elif results[i] == 'Liberal':
      liberal += 1
    elif results[i] == 'NDP':
      ndp += 1 
    elif results[i] == 'Bloc':
      bloc += 1
  
  counted = {'Liberal': liberal, 'Conservative': conservative,
            'NDP': ndp, 'Bloc': bloc}
  
  return counted
      
```

```python
election_results_2015 = pd.read_csv('ResultsbyelectoralDistrict.csv', encoding='Latin1')

list2015 = partylist(election_results_2015)
census_df = census_orig()

results2015_census_df = resultsframe(list2015, census_df)

election_list_df = results2015_census_df

election_list_df.head()
```

```python
#Create the list for 2011
election_results_2011 = pd.read_csv('resultsbyelectoralDistrict2011.csv')

print("Orig List =", election_results_2011.shape)
list2011 = partylist(election_results_2011)
print("Size=", len(list2011))
census_df = census_orig()

results2011_census_df = resultsframe(list2011, census_df)

election_list_df = election_list_df.append(results2011_census_df, ignore_index=True)

print(election_list_df.shape)

election_list_df.head()
```
