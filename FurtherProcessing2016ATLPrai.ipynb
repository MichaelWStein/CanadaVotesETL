```python
import pandas as pd
import json
import numpy as np
```

```python
census_df = pd.read_pickle("Census2016ElectoralDistrictsInfo.pkl")

census_df.head()
```

```python
census_df = census_df.rename(columns={'      English': 'Eng', '      French': 'Fr', '  Non-immigrants': 'Non-Im','  Immigrants': 'Im',
                                      '    Non-official languages': 'Non-of', '  18 to 64 years (%)': 'PovertyRate' })

print(census_df.columns)

df_calc = census_df
```

```python
df_calc = df_calc[df_calc.PROVINCE != 'Quebec']
df_calc = df_calc[df_calc.PROVINCE != 'Ontario']
df_calc = df_calc[df_calc.PROVINCE != 'British Columbia']
df_calc = df_calc[df_calc.PROVINCE != 'Yukon']
df_calc = df_calc[df_calc.PROVINCE != 'Nunavut']
df_calc = df_calc[df_calc.PROVINCE != 'Northwest Territories']

df_calc.reset_index(inplace = True)

census_df = df_calc

census_df.tail()
```

```python
#Reworking language and immigrants to percentage, filling the missing poverty rate with dummies

for i in range(0, census_df.shape[0]):
    a = census_df.loc[i, 'Eng']
    b = census_df.loc[i, 'Fr']
    c = census_df.loc[i, 'Non-of']
    total = a + b + c
    a1 = a / total
    b1 = b / total
    c1 = c / total
    census_df.at[i, 'Eng'] = a1
    census_df.at[i, 'Fr'] = b1
    census_df.at[i, 'Non-of'] = c1
    
    d = census_df.loc[i, 'Non-Im']
    e = census_df.loc[i, 'Im']
    e1 = e / (d + e)
    census_df.at[i, 'Im'] = e1
    
    d = census_df.loc[i, 'PovertyRate']
    if d == None:
        census_df.at[i, 'PovertyRate'] = 20
        print(i)
      

```

```python
# Only one of the characteristics of immigrants / non-immigrants is needed
census_df = census_df.drop(columns = ['Non-Im'])

census_df = census_df.drop(columns = ['index'])


```
